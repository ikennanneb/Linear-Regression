## Simple Linear Regression (Gradient descent)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_dataset.csv")

x = df["YearsExperience"].values

y = df["Salary"].values
```

```python
plt.scatter(x, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
def gd(b, m):
    d_ssr_intercept = d_ssr_slope = 0
    
    for i in range(len(x)):
        residual = (y[i] - (b + m * x[i]))
        d_ssr_intercept += -2 * residual
        d_ssr_slope += -2 * x[i] * residual

    step_size_slope = d_ssr_slope * 0.0001
    step_size_intercept = d_ssr_intercept * 0.0001

    b -= step_size_intercept
    m -= step_size_slope

    return b, m
```

```python
b = count = 0
m = 0.5

while count < 10000:
    b, m = gd(b, m)
    count += 1

X = np.linspace(min(x), max(x), 100)

Y = m * X + b
```

```python
plt.scatter(x, y)
plt.plot(X, Y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

print(f"Best Fit Formula: y = {m:.2f}x + {b:.2f}")
```
