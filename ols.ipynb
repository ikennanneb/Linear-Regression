## Linear Regression (Ordinary Least Squares)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_dataset.csv")

x = df["YearsExperience"].values

y = df["Salary"].values
```

```python
plt.scatter(x, y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
mean_x = sum(x) / len(x)
mean_y = sum(y) / len(y)

m_numerator = [0] * len(y)
m_denominator = [0] * len(y)

for i in range(len(m_numerator)):
    m_numerator[i] = (x[i] - mean_x) * (y[i] - mean_y)
    m_denominator[i] = (x[i] - mean_x)**2

m = sum(m_numerator) / sum(m_denominator)

b = mean_y - m * mean_x

X = np.linspace(min(x), max(x), 100)

Y = m * X + b
```

```python
plt.scatter(x, y)
plt.plot(X, Y)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

print(f"Best Fit Formula: y = {m:.2f}x + {b:.2f}")
```
